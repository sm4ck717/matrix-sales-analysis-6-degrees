## RFM Analysis

```python
task5_df
```

```python
analysis_date = task5_df['sale_date'].max() + pd.Timedelta(days=1)
```

```python
rfm = (
    task5_df
    .groupby('customer_id')
    .agg(
        recency=('sale_date', lambda x: (analysis_date - x.max()).days),
        frequency=('sale_id', 'nunique'),
        monetary=('total_amount', 'sum')
    )
    .reset_index()
)

rfm
```

```python
rfm['R_score'] = pd.qcut(
    rfm['recency'],
    q=5,
    labels=[5, 4, 3, 2, 1]
).astype(int)
```

```python
rfm['F_score'] = pd.qcut(
    rfm['frequency'].rank(method='first'),
    q=5,
    labels=[1, 2, 3, 4, 5]
).astype(int)
```

```python
rfm['M_score'] = pd.qcut(
    rfm['monetary'],
    q=5,
    labels=[1, 2, 3, 4, 5]
).astype(int)
```

```python
def rfm_segment(row):
    if row['R_score'] == 5 and row['F_score'] == 5 and row['M_score'] == 5:
        return 'Top Customers'
    elif row['F_score'] >= 4 and row['M_score'] >= 4:
        return 'Loyal Customers'
    elif row['R_score'] == 5:
        return 'Recent Customers'
    elif row['M_score'] == 5:
        return 'Big Spenders'
    elif row['R_score'] == 1 and row['F_score'] == 1 and row['M_score'] == 1:
        return 'Low Value Customers'
    elif row['R_score'] >= 4 and row['F_score'] >= 3:
        return 'Potential Loyalists'
    elif (row['R_score'] in [2, 3]) and (row['F_score'] in [2, 3]):
        return 'At Risk Customers'
    elif row['R_score'] == 1:
        return 'Lost Customers'
    else:
        return 'Others'
```

```python
rfm['segment'] = rfm.apply(rfm_segment, axis=1)
```

## BCG Matrix

```python
task3_df
```

```python
product_bcg = (
    task3_df
    .groupby(['product_id', 'product_name', 'category'])
    .agg(
        revenue=('item_total', 'sum'),
        units_sold=('quantity', 'sum'),
        avg_margin_pct=('gross_margin_pct', 'mean')
    )
    .reset_index()
)
```

```python
total_revenue = product_bcg['revenue'].sum()
product_bcg['relative_revenue_share'] = (
    product_bcg['revenue'] / total_revenue
)
```

```python
product_bcg['demand_velocity'] = product_bcg['units_sold']
```

```python
rev_threshold = product_bcg['relative_revenue_share'].median()
demand_threshold = product_bcg['demand_velocity'].median()
```

```python
def bcg_class(row):
    if row['relative_revenue_share'] >= rev_threshold and row['demand_velocity'] >= demand_threshold:
        return 'Stars'
    elif row['relative_revenue_share'] >= rev_threshold and row['demand_velocity'] < demand_threshold:
        return 'Cash Cows'
    elif row['relative_revenue_share'] < rev_threshold and row['demand_velocity'] >= demand_threshold:
        return 'Question Marks'
    else:
        return 'Dogs'

product_bcg['bcg_quadrant'] = product_bcg.apply(bcg_class, axis=1)
```

```python
bcg_margin = (
    product_bcg
    .groupby('bcg_quadrant')
    .agg(
        products=('product_id', 'nunique'),
        avg_margin_pct=('avg_margin_pct', 'mean'),
        total_revenue=('revenue', 'sum')
    )
    .reset_index()
)

bcg_margin
```
